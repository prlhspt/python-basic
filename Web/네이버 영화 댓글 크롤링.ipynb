```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd


point = []
review = []
date = []

try :
    for i in range(1, 30):
        url = ('https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=186615&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={}'.format(i))
        html = urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')


        results = soup.find(class_='score_result')
        results

        lis = results.find_all('li')

        for li in lis:
            # 리뷰쓴 사람이 관람객일 경우(1페이지 li[4])
            if (li.find(class_='score_reple').find('p').find('span').get_text().strip()) == '관람객':
                review.append(li.find(class_='score_reple').find('p').find_all('span')[1].get_text().strip())

                
# 개선 필요  ----------------------             
#             # 태그가 1개보다 많을 경우
#             elif len(li.find(class_='score_reple').find('p').find('span')) > 1:


#                 # 리뷰가 길어서 접혀있는경우 (span태그가 2개이상일 경우(접혀있다는 뜻의 class))
#                 if len(li.find(class_='score_reple').find('p').find('span').get_text().strip()) > 50:
#                     review.append(li.find(class_='score_reple').find('p').find('span').find('a')['data-src'])


#                 # 스포일러 방지일 경우
#                 #if (li.find(class_='score_reple').find('p').find('span')['class'][0]) == 'text_spo':
#                 else:
#                     li.find(class_='score_reple').find_all('span')[1].string.strip()
# ---------------------------------



            #아닌경우
            else:
                review.append(li.find(class_='score_reple').find('p').find('span').get_text().strip())

            #multi date
            li.find(class_='score_reple').find_all('em')[0]
            date.append(li.find(class_='score_reple').find_all('em')[1].get_text())

            #multi score
            point.append(li.find(class_='star_score').find('em').string)
except :
    print(i)
    
```
